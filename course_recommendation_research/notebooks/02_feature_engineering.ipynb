## Objectives

- Define a **reproducible feature set** for course success prediction.
- Create modeling targets (`profit`, `log1p(profit)`).
- Save train/test splits for consistency across models.

**Output:** `data/processed/train.parquet`, `data/processed/test.parquet`


```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split

DATA_PATH = Path('..') / 'data' / 'processed' / 'courses_clean.parquet'
df = pd.read_parquet(DATA_PATH)

df['log_profit'] = np.log1p(df['profit'])

def make_features(d: pd.DataFrame) -> pd.DataFrame:
    out = d.copy()
    out['title_len'] = out['course_title'].astype(str).str.len()
    out['content_duration_hours'] = out['content_duration_hours'].fillna(out['content_duration_hours'].median())
    return out

df_fe = make_features(df)
df_fe.shape
```

```python
# Split once for the whole project
SEED = 42
train_df, test_df = train_test_split(df_fe, test_size=0.2, random_state=SEED)

(train_df.shape, test_df.shape)
```

```python
# Save splits
out_dir = Path('..') / 'data' / 'processed'
train_path = out_dir / 'train.parquet'
test_path = out_dir / 'test.parquet'
train_df.to_parquet(train_path, index=False)
test_df.to_parquet(test_path, index=False)
(train_path, test_path)
```
